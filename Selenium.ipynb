Scrapping top 100 film d'apres Allo-ciné :

```python
import time
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
browser = webdriver.Chrome()

browser.get('https://www.allocine.fr/film/meilleurs/')
conditions = browser.find_element_by_xpath('//*[@id="cmp-main"]/button[2]/span')
conditions.click()
time.sleep(3)
```

```python
import re
import pandas as panda



df = panda.DataFrame(columns=['Titre','Réalisateur','Acteurs principaux','Genre','Date de parution en salle','Durée'])



#recuêrer les liens des pages suivantes (10 pages)
list_of_links_pages = browser.find_element_by_xpath('//*[@id="content-layout"]/section[3]/div[2]/nav/div') 
links_Pages = []
links_Pages = list_of_links_pages.find_elements_by_tag_name('a')
linkPages = []
for link in links_Pages:
    linkPages.append(link.get_attribute('href'))
#supprimer les liens vers la page 20 et 30 
del linkPages[-1]
del linkPages[-1]
linkPages.insert(10,'https://www.allocine.fr/film/meilleurs/?page=10')

for page in linkPages:
    #recuperer les liens des films 
    list_of_links_movies = browser.find_elements_by_css_selector(".meta-title-link")
    links_movies = []

    for lis in list_of_links_movies:
        links_movies.append(lis.get_attribute('href'))
    
    # Loop through all the links and launch one by one
    for link in links_movies:
        browser.get(link)
        # Scrape here
        #Get all info for one movie ('Titre','Réalisateur','Acteurs principaux','Genre','Date de parution en salle','Durée')

        title = browser.find_element_by_xpath('//*[@id="content-layout"]/div[2]/div[1]').text
        #try & except permet de gerer si dans la page si il y a une bande annance car cela va changer le x path de notre element
        try:
            real = browser.find_element_by_xpath('//*[@id="content-layout"]/section/div/div[3]/div[1]/div/div[3]/a').text
        except:
            real = browser.find_element_by_xpath('//*[@id="content-layout"]/section/div/div[2]/div[1]/div/div[3]/a').text        
        try:
            div_act = browser.find_element_by_xpath('//*[@id="content-layout"]/section/div/div[3]/div[1]/div/div[5]')
        except:
            try:
                 div_act = browser.find_element_by_xpath('//*[@id="content-layout"]/section/div/div[2]/div[1]/div/div[5]')  
            except:
                try:
                    div_act = browser.find_element_by_xpath('//*[@id="content-layout"]/section/div/div[2]/div[1]/div/div[4]')
                except:
                    div_act = browser.find_element_by_xpath('//*[@id="content-layout"]/section/div/div[3]/div[1]/div/div[4]')   
        
        acteurs = div_act.text
        string = "Titre original"
        if string in acteurs:
            try:
                div_act = browser.find_element_by_xpath('//*[@id="content-layout"]/section/div/div[2]/div[1]/div/div[4]')
            except:
                div_act = browser.find_element_by_xpath('//*[@id="content-layout"]/section/div/div[3]/div[1]/div/div[4]')

            acteurs= div_act.text

        acteurs = acteurs.replace("Avec ","")
        acteurs = acteurs.replace("Avec ","")

        div = browser.find_element_by_xpath('//*[@id="content-layout"]/section/div/div[3]/div[1]/div/div[1]')
        text = div.text
        if text == '':
            div = browser.find_element_by_xpath('//*[@id="content-layout"]/section/div/div[2]/div[1]/div/div[1]') 
        text = div.text
        new_text = re.split(' / ', text)

        date = new_text[0]
        date = date.replace(" en salle","")
        dure = new_text[1]
        genre = new_text[2]

        
        print(title,';',real,';',acteurs,';',genre,';',date,';',dure)

        #appendinfo in dataframe 
        df=df.append({'Titre' : title , 'Réalisateur' : real, 'Acteurs principaux' : acteurs, 'Genre' : genre, 'Date de parution en salle' : date,  'Durée' : dure } , ignore_index=True)
    browser.get(page)

        


```
